```python
# Task 1 Count levels.py
from collections import Counter

log_levels = ["DEBUG", "INFO", "WARNING", "ERROR", "CRITICAL"]
counts = Counter()

with open("/Users/edu-loaner/Desktop/log level/app.log", "r") as f:
    for line in f:
        for level in log_levels:
            if level in line:
                counts[level] += 1
                break

for level, count in counts.items():
    print(f"{level}: {count}")
```

```python
# Task 2: Create a new file with only logs mentioning a given user
user_to_find = "Freddy"  # change this as needed

with open("/Users/edu-loaner/Desktop/log level/app.log", "r") as infile, open(f"/Users/edu-loaner/Desktop/log level/{user_to_find}_logs.txt", "w") as outfile:
    for line in infile:
        if user_to_find.lower() in line.lower():
            outfile.write(line)

print(f"Filtered logs written to user_logs.txt for user: {user_to_find}")
#!/usr/bin/env python3
```

```python
# Task 3: Display only ERROR and CRITICAL messages

with open("/Users/edu-loaner/Desktop/log level/app.log", "r") as file:
    for line in file:
        if "ERROR" in line or "CRITICAL" in line:
            print(line.strip())
```

```python
import re

# Task 4: Extract all timestamps using a regex
# Assuming timestamps look like: 2025-10-30 10:12:01
timestamp_pattern = r"\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}"

timestamps = []

with open("/Users/edu-loaner/Desktop/log level/app.log", "r") as file:
    for line in file:
        match = re.search(timestamp_pattern, line)
        if match:
            timestamps.append(match.group())

print("Timestamps found:")
for t in timestamps:
    print(t)

print(f"\nTotal timestamps found: {len(timestamps)}")
```

```python
import re

# Task 5: Find all IP addresses in the log
ip_pattern = r"\b\d{1,3}(?:\.\d{1,3}){3}\b"

with open("/Users/edu-loaner/Desktop/log level/app.log", "r") as file:
    log_data = file.read()

ips = re.findall(ip_pattern, log_data)

print("IP addresses found:")
for ip in set(ips):  # use set() to remove duplicates
    print(ip)

print(f"\nTotal unique IPs: {len(set(ips))}")
```
