```python
# ----------------------------------------
# My Daily Digital Routine Automation
# ----------------------------------------
# Features:
# 1. Opens search engine & Tucson weather URLs, refreshes every 2 minutes.
# 2. Displays real-time clock & calendar (refreshes every 12 hours).
# 3. Performs automatic system cleanup (deletes cache regularly).
# 4. Opens Spotify music playlist.
# ----------------------------------------

import os
import time
import webbrowser
import datetime
import calendar
import shutil

# ----------------------------------------
# 1. URL Automation
# ----------------------------------------

def open_urls():
    print("\n[Automation] Opening search engine and Tucson weather...")
    webbrowser.open("https://www.google.com")
    webbrowser.open("https://weather.com/weather/today/l/Tucson+AZ")
    print("[Automation] URLs opened successfully!")

# ----------------------------------------
# 2. Real-Time Clock & Calendar
# ----------------------------------------

def show_time_and_calendar():
    now = datetime.datetime.now()
    current_time = now.strftime("%I:%M:%S %p")
    current_date = now.strftime("%A, %B %d, %Y")
    
    print("\nüïí Current Time:", current_time)
    print("üìÖ Date:", current_date)
    print("Calendar for this month:\n")
    
    # Highlight today with a green dot (üü¢)
    month_calendar = calendar.monthcalendar(now.year, now.month)
    for week in month_calendar:
        for day in week:
            if day == 0:
                print("   ", end=" ")
            elif day == now.day:
                print(f"{day:2}üü¢", end=" ")
            else:
                print(f"{day:2}", end=" ")
        print()
    print("\n[Automation] Calendar refreshed successfully!")

# ----------------------------------------
# 3. System Cleanup (Cache Deletion)
# ----------------------------------------

def clean_cache():
    print("\nüßπ Cleaning system cache...")
    cache_paths = [
        os.path.expanduser("~\\AppData\\Local\\Temp"),  # Windows
        "/tmp",  # Mac/Linux
    ]
    for path in cache_paths:
        if os.path.exists(path):
            try:
                for filename in os.listdir(path):
                    file_path = os.path.join(path, filename)
                    if os.path.isfile(file_path):
                        os.remove(file_path)
                    elif os.path.isdir(file_path):
                        shutil.rmtree(file_path, ignore_errors=True)
                print(f"‚úÖ Cache cleaned at {path}")
            except Exception as e:
                print(f"‚ö†Ô∏è Could not clean cache at {path}: {e}")
    print("[Automation] System cleanup completed!")

# ----------------------------------------
# 4. Spotify Playlist
# ----------------------------------------

def open_spotify_playlist():
    print("\nüéµ Opening your Spotify playlist...")
    playlist_url = "https://open.spotify.com/playlist/37i9dQZF1DXcBWIGoYBM5M"  # Replace with your playlist link
    webbrowser.open(playlist_url)
    print("[Automation] Enjoy your music!")

# ----------------------------------------
# Scheduling
# ----------------------------------------

# Open URLs every 2 minutes
while True:
    open_urls()
    time.sleep(120)  # 2 minutes

# Refresh calendar every 12 hours
while True:
    show_time_and_calendar()
    time.sleep(43200)  # 12 hours

# Clean cache every 6 hours (adjust as needed)
while True:
    clean_cache()
    time.sleep(21600)  # 6 hours   

# Play Spotify once when program starts
open_spotify_playlist()

# Initial calendar display
show_time_and_calendar()

print("\n‚úÖ Automation System Started ‚Äî Running Tasks Automatically...\n")

# Run the scheduler indefinitely
while True:
    time.sleep(1)
```
